```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import datetime
from datetime import date
```

```python
today = date.today()
start_date= today - datetime.timedelta(days=365)
ticker = "ANX"
```

```python
data = pdr.get_data_yahoo(ticker, start=start_date, end=today)
```

```python
data['SMA30'] =  data['Adj Close'].rolling(window=30).mean()
data['SMA100'] = data['Adj Close'].rolling(window=100).mean()
```

```python
SMA30 = data['SMA30'].array[-1]
SMA100 = data['SMA100'].array[-1]
price = data['Adj Close'].array[-1]
date = data.index.array[-1]
stock = {"ticker" : ticker, "lastSMA30" : SMA30, "lastSMA100": SMA100, "lastPrice": price, "date": date}
stock
```

```python
plt.figure(figsize=(20, 10))
plt.plot(data['Adj Close'], label=ticker, alpha=0.35)
plt.plot(data['SMA30'], label=SMA30, alpha=0.35)
plt.plot(data['SMA100'], label=ticker, alpha=0.35)
plt.title('Adj '+ ticker +' Price with Buy and Sell Signals')
plt.xlabel('Days')
plt.ylabel('Price in $ or ')
plt.xticks(rotation=90)
plt.legend(loc='upper left')
plt.show()
```
